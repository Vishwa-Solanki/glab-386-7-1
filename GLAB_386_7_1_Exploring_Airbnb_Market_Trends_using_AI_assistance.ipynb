````python
# Import necessary packages such as pandas and numpy
import pandas as pd
import numpy as np
````

````python
# Import CSV for prices
airbnb_price = pd.read_csv('airbnb_price.csv') #

# Import Excel file for room types
airbnb_room_type = pd.read_excel('airbnb_room_type.xlsx')

# Import TSV for review dates
airbnb_last_review = pd.read_csv('airbnb_last_review.tsv', sep='\t')
````

````python
# Step 1: Merge airbnb_price and airbnb_room_type on 'listing_id'
# We use an 'inner' merge because we confirmed that all listing_ids are present in both datasets.
airbnb_merged = pd.merge(airbnb_price, airbnb_room_type, on='listing_id', how='inner')
print("--- After first merge (price + room_type) ---")
print(f"Shape of merged DataFrame: {airbnb_merged.shape}\n")

# Step 2: Merge the result with airbnb_last_review on 'listing_id'
# Again, an 'inner' merge is appropriate as 'listing_id' is consistent across all datasets.
airbnb_final = pd.merge(airbnb_merged, airbnb_last_review, on='listing_id', how='inner')
print("--- After second merge (price + room_type + last_review) ---")
print(f"Shape of final DataFrame: {airbnb_final.shape}\n")

print("--- First 5 rows of the final merged DataFrame ---")
print(airbnb_final.head())
````

````python
# Clean the 'price' column: remove ' dollars' and convert to numeric
airbnb_final['price'] = airbnb_final['price'].str.replace(' dollars', '', regex=False)
airbnb_final['price'] = pd.to_numeric(airbnb_final['price'])

# Calculate the average listing price
average_price = airbnb_final['price'].mean()

# Print the result clearly
print(f"The average listing price is: ${average_price:.2f}")
````

**Prompts for Descriptive pricing analysis**


- What’s the overall price distribution (median, IQR, tails)?

- Which room types are most/least expensive?

- Are there obvious outliers to cap or investigate?

````python
# Overall Price Distribution (median, IQR, tails)
print("\n--- Overall Price Distribution ---")
price_median = airbnb_final['price'].median()
Q1 = airbnb_final['price'].quantile(0.25)
Q3 = airbnb_final['price'].quantile(0.75)
IQR = Q3 - Q1

print(f"Median Price: ${price_median:.2f}")
print(f"Q1 (25th percentile): ${Q1:.2f}")
print(f"Q3 (75th percentile): ${Q3:.2f}")
print(f"IQR (Interquartile Range): ${IQR:.2f}")

# Describe the tails (e.g., 5th and 95th percentiles, min/max)
price_5th_percentile = airbnb_final['price'].quantile(0.05)
price_95th_percentile = airbnb_final['price'].quantile(0.95)
price_min = airbnb_final['price'].min()
price_max = airbnb_final['price'].max()

print(f"5th Percentile Price: ${price_5th_percentile:.2f}")
print(f"95th Percentile Price: ${price_95th_percentile:.2f}")
print(f"Minimum Price: ${price_min:.2f}")
print(f"Maximum Price: ${price_max:.2f}")

# Room types most/least expensive
print("\n--- Price by Room Type ---")
room_type_prices = airbnb_final.groupby('room_type')['price'].agg(['mean', 'median', 'min', 'max'])
print(room_type_prices.sort_values(by='mean', ascending=False))

# Outlier detection using IQR method
# Define bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = airbnb_final[(airbnb_final['price'] < lower_bound) | (airbnb_final['price'] > upper_bound)]

print(f"\n--- Outlier Analysis ---")
print(f"Lower Bound for Outliers: ${lower_bound:.2f}")
print(f"Upper Bound for Outliers: ${upper_bound:.2f}")
print(f"Number of outliers: {len(outliers)}")
print("First 5 outliers (if any):\n")
print(outliers.head())
````

**Propmts for Inventory mix & positioning**


- What share of listings are Entire home/apt vs Private room vs others?

 - Do some room types dominate certain price bands?

````python
# Part 1: Share of listings by room type
print("\n--- Share of Listings by Room Type ---")
room_type_distribution = airbnb_final['room_type'].value_counts(normalize=True) * 100
print(room_type_distribution.round(2).astype(str) + '%')

# Part 2: Room types dominating certain price bands
print("\n--- Room Type Distribution Across Price Bands ---")

# Create price bands using quantiles for a more balanced distribution
# Exclude price 0 to avoid issues with quantile calculation and very low outliers
price_data_for_bands = airbnb_final[airbnb_final['price'] > 0]['price']

if not price_data_for_bands.empty:
    airbnb_final['price_band'] = pd.qcut(
        airbnb_final['price'],
        q=4, # Create 4 quartiles: Low, Medium, High, Very High
        labels=['Low Price', 'Medium Price', 'High Price', 'Very High Price'],
        duplicates='drop' # Handle cases with identical bin edges due to many zero prices
    )

    # Analyze room type distribution within these price bands
    # Count listings per room type in each price band
    room_type_price_band_counts = airbnb_final.groupby('price_band')['room_type'].value_counts().unstack(fill_value=0)

    # Calculate percentages within each price band
    room_type_price_band_percentage = room_type_price_band_counts.apply(lambda x: x / x.sum() * 100, axis=1)

    print("Counts of Room Types per Price Band:\n")
    print(room_type_price_band_counts)
    print("\nPercentage Distribution of Room Types within Each Price Band:\n")
    print(room_type_price_band_percentage.round(2).astype(str) + '%')
else:
    print("Cannot create price bands as all prices are zero or data is empty.")
````

**Propmt for Review recency & activity**

- How recent are reviews? Is demand cooling/heating?

- Are there seasonal peaks in last reviews?

````python
# Convert 'last_review' in airbnb_final to datetime, if not already converted
# It appears from the kernel state that 'last_review' in airbnb_final is still a string.
# The earlier conversion was on airbnb_last_review, not the merged airbnb_final.
if pd.api.types.is_string_dtype(airbnb_final['last_review']):
    airbnb_final['last_review'] = pd.to_datetime(airbnb_final['last_review'])

print("\n--- Review Recency Analysis ---")
# Calculate the date of the most recent review
most_recent_review = airbnb_final['last_review'].max()
print(f"The most recent review date is: {most_recent_review.strftime('%Y-%m-%d')}")

# Calculate the overall date range of reviews
min_review_date_final = airbnb_final['last_review'].min()
max_review_date_final = airbnb_final['last_review'].max()
print(f"The overall review date range is from {min_review_date_final.strftime('%Y-%m-%d')} to {max_review_date_final.strftime('%Y-%m-%d')}")

# To assess 'demand cooling/heating', we would ideally need a time series of new listings or bookings.
# With 'last_review' only, we can infer activity by looking at the distribution of these dates.
# For example, if many reviews are recent, activity is likely high.

# --- Seasonal Peaks in Last Reviews ---
print("\n--- Seasonal Review Activity ---")
# Extract month from last_review and count reviews per month
airbnb_final['review_month'] = airbnb_final['last_review'].dt.month_name()
monthly_reviews = airbnb_final['review_month'].value_counts().sort_index()

# Sort by month order for better readability
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
monthly_reviews = monthly_reviews.reindex(month_order).dropna().astype(int)

print("Number of reviews by month:\n")
print(monthly_reviews)

# Percentage of reviews by month
monthly_reviews_percent = (airbnb_final['review_month'].value_counts(normalize=True) * 100).reindex(month_order).dropna().round(2)
print("\nPercentage of reviews by month:\n")
print(monthly_reviews_percent.astype(str) + '%')
````

**Propmts for Market segmentation**

- Are there price differences across room-type segments?

- If we have neighborhood/borough (not seen here), how does location shift price?

````python
# Part 1: Price differences across room-type segments (re-iterating from previous analysis)
print("\n--- Price Differences Across Room Types (Mean, Median, Min, Max) ---")
room_type_prices = airbnb_final.groupby('room_type')['price'].agg(['mean', 'median', 'min', 'max'])
print(room_type_prices.sort_values(by='mean', ascending=False))

# Part 2: How location (neighborhood/borough) shifts price
print("\n--- Price by Neighborhood/Borough ---")
# Calculate the average price for each neighborhood/borough
neighborhood_prices = airbnb_final.groupby('nbhood_full')['price'].mean().sort_values(ascending=False)

print("Top 10 most expensive neighborhoods/boroughs (average price):\n")
print(neighborhood_prices.head(10))

print("\nTop 10 least expensive neighborhoods/boroughs (average price):\n")
print(neighborhood_prices.tail(10))
````
